```python
lens_folders = glob.glob("../lens_systems/*")

system_names = [
    "DESIJ0132-1600",
    "DESIJ0136-0008",
    "DESIJ0201-2739",
    "DESIJ0215-2909",
    "DESIJ0618+5018",
    "DESIJ0923+1829",
    "DESIJ1018-0121",
    "DESIJ1101-0602",
    "DESIJ1205+4110",
    "DESIJ1501+5208",
    "DESIJ1537+1443",
    "DESIJ1554+0443",
    "DESIJ1624+0129",
    "DESIJ1709+3154",
    "DESIJ2158+0257",
]

len(system_names)
```

```python
def load_image_data(system_name):
    """
    Load image data from HDF5 file.

    :param system_name: name of the lens system
    :type system_name: str
    :return: image data
    :rtype: numpy.ndarray
    """
    file_path = f"../lens_systems/{system_name}/{system_name}_F140W.h5"

    with h5py.File(file_path, "r") as f:
        kwargs_data = {key: f[key][()] for key in f}

    return kwargs_data["image_data"]


def load_transform_pix2angle(system_name):
    """
    Load transform_pix2angle from HDF5 file.

    :param system_name: name of the lens system
    :type system_name: str
    :return: transform_pix2angle
    :rtype: numpy.ndarray
    """
    file_path = f"../lens_systems/{system_name}/{system_name}_F140W.h5"

    with h5py.File(file_path, "r") as f:
        kwargs_data = {key: f[key][()] for key in f}

    return kwargs_data["transform_pix2angle"]


def load_def_redshift_values(system_name):
    """
    Load redshift values from yml file.

    :param system_name: name of the lens system
    :type system_name: str
    :return: deflector redshift values
    :rtype: numpy.ndarray
    """
    file_path = f"../lens_systems/{system_name}/{system_name}_point_estimates.yml"

    try:
        with open(file_path, "r") as f:
            data = yaml.full_load(f)

    except AttributeError:
        print(f"Failed to load {system_name} redshift values.")
        return None

    return data["z_phot"]


def load_src_redshift_values(system_name):
    """
    Load redshift values from yml file.

    :param system_name: name of the lens system
    :type system_name: str
    :return: source redshift values
    :rtype: numpy.ndarray
    """
    file_path = f"../lens_systems/{system_name}/{system_name}_point_estimates.yml"

    try:
        with open(file_path, "r") as f:
            data = yaml.full_load(f)

    except AttributeError:
        print(f"Failed to load {system_name} redshift values.")
        return None
    # If there is no source redshift, return None
    try:
        return data["z_src"]
    except KeyError:
        return None
```

```python
def plot_one_system(ax, system_name, text_arc=None, z_def=None, z_src=None):
    """
    Plot image of one lens system.

    :param system_name: name of the lens system
    :type system_name: str
    :return: None
    :rtype:
    """
    image_data = load_image_data(system_name)
    image_data[image_data < 0] = np.min(np.abs(image_data))

    ax.matshow(np.log10(image_data), vmin=-6, vmax=1.5, cmap=cmap, origin="lower")
    ax.set_xticks([])
    ax.set_yticks([])

    # Adjust the position and alignment of the text
    # Print except for the second system
    if system_name != "DESIJ0136-0008":
        ax.text(
            0.13,
            1.1,
            system_name,
            transform=ax.transAxes,
            fontsize=20,  # Adjust as per `pf.mnras_text_fontsize`
            color="black",
            va="top",
            ha="left",
            # bbox=dict(facecolor="k", alpha=0.90, edgecolor="none"),
        )

    # Edit the second system name and delete the first one
    if system_name == "DESIJ0136-0008":
        ax.text(
            0.13,
            1.1,
            "DESIJ0136+0008",
            transform=ax.transAxes,
            fontsize=20,  # Adjust as per `pf.mnras_text_fontsize`
            color="black",
            va="top",
            ha="left",
            # bbox=dict(facecolor="k", alpha=0.90, edgecolor="none"),
        )

    # Add redshift value at top left corner
    ax.text(
        0.05,
        0.95,
        "$z_{{def}} = {:.3f}$".format(
            z_def[0]
        ),  # or including err "$z_{{def}} = {} \pm {}$".format(z_def[0], z_def[1]),
        transform=ax.transAxes,
        fontsize=20,
        color="white",
        va="top",
        ha="left",
        # bbox=dict(facecolor="k", alpha=0.90, edgecolor="none"),
    )

    # If there is a z_src value, add it to the plot
    if z_src is not None:
        ax.text(
            0.05,
            0.85,
            # Print upto 1 decimal places
            "$z_{{src}} = {:.3f}$".format(z_src[0]),
            # or including err "$z_{{src}} = {} \pm {}$".format(z_def[1], z_def[2]),
            transform=ax.transAxes,
            fontsize=20,
            color="yellow",
            va="top",
            ha="left",
            # bbox=dict(facecolor="k", alpha=0.90, edgecolor="none"),
        )

    t_p2a = load_transform_pix2angle(system_name)

    # Calculate the length of the ruler in pixels
    arcsec_per_pixel = np.diag(t_p2a)[0]
    ruler_length_arcsec = -1 / arcsec_per_pixel  # Length of the scale bar in arcseconds

    # Add scale bar using lenstronomy's utility function
    plot_util.scale_bar(
        ax, 50, dist=ruler_length_arcsec, text=" {}".format(text_arc), font_size=25
    )
```

```python
num_col = 5
num_row = (len(system_names) - 1) // num_col + 1
save_figure = True

fig, axes = plt.subplots(
    num_row, num_col, figsize=(20, 4 * num_row)
)  # Adjust figure size as necessary

for i, ax in enumerate(axes.flatten()):
    if i == 0:
        plot_one_system(
            ax,
            system_names[i],
            text_arc='1"',
            z_def=load_def_redshift_values(system_names[i]),
            z_src=load_src_redshift_values(system_names[i]),
        )
    if i < len(system_names):
        plot_one_system(
            ax,
            system_names[i],
            text_arc="",
            z_def=load_def_redshift_values(system_names[i]),
            z_src=load_src_redshift_values(system_names[i]),
        )
    else:
        ax.axis("off")

fig.subplots_adjust(hspace=0.2, wspace=0.00)

if save_figure:
    fig.savefig("../paper/figures/lens_montage.pdf", bbox_inches="tight")

plt.show()
```
